```python
import csv
import matplotlib.pyplot as plt
import numpy as np

n_cells = 3000
cell_sums = []
for i in range(0,n_cells):
    with open('simulations/E'+str(i)+'.csv') as f:
        reader = csv.reader(f)
        next(reader)
        sum_cell = 0
        k= 0
        for ln in reader:
            if k == 2:
                sum_cell += float(ln[len(ln)-1])
            k+=1
    cell_sums.append(sum_cell)

print(len(cell_sums))
plt.hist(cell_sums)
```

```python
import numpy as np
print(cell_sums)
cs = np.array(cell_sums)
clusters = (cs>0.7)*np.ones(len(cs))

np.savetxt("clusters.csv", clusters, delimiter=",")
```
